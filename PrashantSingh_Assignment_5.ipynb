**Function For Gradient Descent**

* Minimizes a given objective function by iteratively updating parameters based on the gradient.

* Takes an initial guess (start_x), a learning rate (learning_rate), and the number of iterations (num_iterations).

* Computes the gradient, updates the parameters by moving in the opposite direction of the gradient, and stores the objective function values for each iteration.

* Returns the optimized parameters, the final objective function value, and a history of values for analysis.

```python
def gradient_descent(start_x, learning_rate, num_iterations):
    x = start_x
    history = []

    for i in range(num_iterations):
        # Calculate the gradient
        grad_x = df_dx(x)

        # Update the parameter
        x = x - learning_rate * grad_x

        # Save the objective function value
        history.append(f(x))


    return x, f(x), history
```

**Question-2(b)**

**Solution**

Defining function min x1^2 + x2^2 + 4x2 + 3x1 + 7 and finding 1st Partial derivative.

```python
def f_b(x1, x2):
    return x1**2 + x2**2 + 4 * x2 + 3 * x1 + 7

def df_dx_b1(x1, x2):
    return 2 * x1 + 3

def df_dx_b2(x1, x2):
    return 2 * x2 + 4
```

```python
def gradient_descent_b(start_x1, start_x2, learning_rate, num_iterations):
    x1 = start_x1
    x2 = start_x2
    history = []

    for i in range(num_iterations):
        grad_x1 = df_dx_b1(x1, x2)
        grad_x2 = df_dx_b2(x1, x2)

        x1 = x1 - learning_rate * grad_x1
        x2 = x2 - learning_rate * grad_x2

        history.append(f_b(x1, x2))

    return x1, x2, f_b(x1, x2), history

```

```python
start_x1_b = 0.0
start_x2_b = 0.0
num_iterations_b = 100
```

```python
learning_rates_b = [0.1, 0.01, 0.001, 0.0001]

results_b = []

for learning_rate_b in learning_rates_b:
    final_x1_b, final_x2_b, final_value_b, history_b = gradient_descent_b(start_x1_b, start_x2_b, learning_rate_b, num_iterations_b)
    results_b.append((learning_rate_b, final_x1_b, final_x2_b, final_value_b, history_b))

    print(f"Results for Part (b) with Learning Rate: {learning_rate_b}")
    print(f"Final x1: {final_x1_b}, Final x2: {final_x2_b}, Final Objective Function Value: {final_value_b}\n")

    plt.plot(history_b, label=f'LR: {learning_rate_b}')
```

```python
plt.figure(figsize=(8, 4))
plt.plot(history_b, label='Objective Function Value')
plt.title('Objective Function Value vs Iterations (Part b)')
plt.xlabel('Iteration')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()
```
